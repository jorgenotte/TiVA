```python
# Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import eurostat
```

```python
# Read the table of contents and return a list of tuples (Specific table)
toc = eurostat.get_toc(dataset='naio_10_fcp_ii3', lang='en')
toc
```

```python
# Read the table of contents of the main database and return a dataframe
toc_df = eurostat.get_toc_df()
toc_df
```

```python
# To get only the metadata of one dataset
toc_sp = eurostat.get_toc_df(dataset='naio_10_fcp_ii3', lang='en')
toc_sp
```

```python
# Read the parameter names that can be filtered for a given dataset code and return them as a list.
pars = eurostat.get_pars('naio_10_fcp_ii3')
pars
```

```python
# Read the values of a given parameter par that can be found in a given dataset code
values = eurostat.get_par_values('naio_10_fcp_ii3', 'ind_ava')
values
```

```python
# Get the dictionary of the the parameter values of a dataset as a dataframe
dic = eurostat.get_dic('naio_10_fcp_ii3', frmt='df')
dic
```

```python
# Full download a dataset without flags as pandas dataframe
data = eurostat.get_data_df('naio_10_fcp_ii3')
data
```

```python
# Download a subset of data from a dataset without flags as pandas dataframe
code = 'naio_10_fcp_ii3'
my_filter_pars = {'c_orig': 'ES', 'c_dest': 'FR', 'ind_use': ['A01','A02','A03'], 'ind_ava': ['A01','A02','A03']}
data = eurostat.get_data_df(code, filter_pars=my_filter_pars)
data
```
