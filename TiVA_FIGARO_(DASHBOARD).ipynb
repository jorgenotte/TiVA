```python
# Leer el fichero TiVA en un dataframe [FORMATO CORTO]
ruta1='/content/drive/MyDrive/FIGARO/DATASETS/flatfile_eu-ic-io_ind-by-ind_23ed_2018.csv'
ruta2='/content/drive/MyDrive/FIGARO/DATASETS/flatfile_eu-ic-io_ind-by-ind_23ed_2019.csv'
ruta3='/content/drive/MyDrive/FIGARO/DATASETS/flatfile_eu-ic-io_ind-by-ind_23ed_2020.csv'
ruta4='/content/drive/MyDrive/FIGARO/DATASETS/flatfile_eu-ic-io_ind-by-ind_23ed_2021.csv'

df_2018=pd.read_csv(ruta1)
df_2019=pd.read_csv(ruta2)
df_2020=pd.read_csv(ruta3)
df_2021=pd.read_csv(ruta4)
```

```python
df_2018['year'] = 2018
df_2019['year'] = 2019
df_2020['year'] = 2020
df_2021['year'] = 2021
```

```python
df = pd.concat([df_2018, df_2019, df_2020, df_2021], ignore_index=True)
del [df_2018, df_2019, df_2020, df_2021]
```

```python
# Se definen las variables del VA
valores1 = ['D21X31', 'OP_NRES', 'OP_RES', 'D1', 'D29X39', 'B2A3G']

# Se definen las variables de la demanda final
valores2 = ['P3_S13', 'P3_S14', 'P3_S15', 'P51G', 'P5M']

# Expresión regular que coincide con cualquiera de los caracteres exactos especificados
patron1 = '|'.join(valores1)
patron2 = '|'.join(valores2)
```

```python
df
```

```python
# Se crea el siguiente df: Consumos intermedios + Demanda final
df_z = df[~df['rowIi'].str.contains(patron1, regex=True)].reset_index(drop=True)
del [df]
```

```python
df_z
```

```python
df_z['aux'] = (df_z['refArea'] == df_z['counterpartArea']).astype(int)
```

```python
df_z
```

```python
df_z = df_z[df_z['aux'] == 0]
```

```python
df_z_exp = pd.DataFrame(df_z.groupby(['refArea','year'])['obsValue'].sum()).reset_index()
df_z_exp = df_z_exp.rename(columns={'refArea': 'country'})
```

```python
df_z_imp = pd.DataFrame(df_z.groupby(['counterpartArea','year'])['obsValue'].sum()).reset_index()
df_z_imp = df_z_imp.rename(columns={'counterpartArea': 'country'})
```

```python
resultado_consulta = df_z_exp[df_z_exp['country'] == 'ES']
```

```python
resultado_consulta1 = df_z_imp[df_z_imp['country'] == 'ES']
```

```python
df_z_ratio = df_z_exp.copy()
df_z_ratio = df_z_ratio.rename(columns={'refArea': 'country'})
df_z_ratio = df_z_ratio.drop(columns=['obsValue'])
```

```python
df_z_ratio = df_z_ratio.assign(ratio=df_z_exp['obsValue'] / df_z_imp['obsValue'])
```

```python
ruta_excel = '/content/drive/MyDrive/FIGARO/DASHBOARD/Dashboard.xlsx'

with pd.ExcelWriter(ruta_excel, engine='openpyxl') as writer:
    # Guardar cada DataFrame en una hoja diferente
    df_z_exp.to_excel(writer, sheet_name='Exports', index=False)
    df_z_imp.to_excel(writer, sheet_name='Imports', index=False)
    df_z_ratio.to_excel(writer, sheet_name='Ratio exp-imp', index=False)
```
